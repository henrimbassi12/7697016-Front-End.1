```python
# üì¶ Imports de base n√©cessaires pour cette √©tape
import pandas as pd
from google.colab import files
import os # Utile si on bascule sur Google Drive plus tard

# --- M√©thode pour Uploader le fichier Excel (simple pour commencer) ---
# Sp√©cifiez le nom exact du fichier Excel √† uploader
file_to_upload = "RAPPORT D'ENTRETIEN FRIGO AVRIL-JUIN.xlsx"

print(f"Uploader le fichier Excel {file_to_upload}...")
try:
    # Demande √† l'utilisateur d'uploader le fichier
    uploaded = files.upload()

    # V√©rifier si le fichier attendu est dans le dictionnaire des fichiers upload√©s
    if file_to_upload in uploaded:
        print(f"‚úÖ Fichier Excel '{file_to_upload}' upload√© avec succ√®s.")

        # Lire le fichier Excel dans un DataFrame pandas
        # Utiliser engine='openpyxl' est recommand√© pour les fichiers .xlsx
        df = pd.read_excel(file_to_upload, engine='openpyxl')
        print("‚úÖ Donn√©es charg√©es dans un DataFrame pandas.")

    else:
        # Si le dictionnaire n'est pas vide mais le fichier attendu n'est pas l√†
        if uploaded:
             print(f"‚ùå Erreur : Vous avez upload√© un autre fichier (ou plusieurs).")
             print(f"Fichier(s) upload√©(s) : {list(uploaded.keys())}")
             print(f"Le fichier attendu √©tait : '{file_to_upload}'")
        else:
            # Si le dictionnaire est vide (upload annul√© ou √©chec)
            print(f"‚ùå Erreur : L'upload du fichier '{file_to_upload}' a √©chou√© ou a √©t√© annul√©.")

        df = None # Assigner None si le fichier attendu n'est pas charg√©

except FileNotFoundError:
    print(f"‚ùå Erreur : Le fichier upload√© '{file_to_upload}' n'a pas √©t√© trouv√© apr√®s l'upload.")
    df = None
except Exception as e:
    print(f"‚ùå Erreur lors de l'upload ou du chargement du fichier Excel : {e}")
    df = None # Assigner None en cas d'erreur


# --- Afficher les premi√®res informations sur les donn√©es charg√©es ---
if df is not None:
    print("\n--- Informations sur les donn√©es charg√©es ---")
    print("Premi√®res lignes :")
    display(df.head()) # Utilisez display() pour un affichage plus propre dans Colab
    print("\nInformation sur les colonnes :")
    df.info()
    print("\nR√©sum√© statistique des colonnes num√©riques :")
    display(df.describe()) # Utilisez display() ici aussi
    print("\nNoms des colonnes pour r√©f√©rence :")
    print(df.columns.tolist())
else:
    print("\nImpossible de charger les donn√©es. Veuillez v√©rifier le nom du fichier ('RAPPORT D'ENTRETIEN FRIGO AVRIL-JUIN.xlsx') et r√©essayer.")

# Marquez ici si le chargement a r√©ussi pour les √©tapes suivantes
df_loaded = df is not None
```

```python
# √âtape 2 : Exploration et Nettoyage des donn√©es

if df_loaded: # S'assurer que le DataFrame a bien √©t√© charg√© √† l'√©tape pr√©c√©dente
    print("\n--- D√©but de l'exploration et du nettoyage des donn√©es ---")

    # 2.1 V√©rifier les types de donn√©es
    print("\nTypes de donn√©es de chaque colonne :")
    print(df.dtypes)

    # 2.2 V√©rifier les valeurs manquantes
    print("\nNombre de valeurs manquantes par colonne :")
    print(df.isnull().sum())

    # 2.3 Nombre de lignes et colonnes
    print(f"\nLe DataFrame contient {df.shape[0]} lignes et {df.shape[1]} colonnes.")

    # 2.4 Afficher un √©chantillon al√©atoire pour avoir une id√©e rapide des donn√©es
    print("\n√âchantillon al√©atoire des donn√©es :")
    display(df.sample(5)) # Affiche 5 lignes al√©atoires

else:
    print("\nImpossible de proc√©der √† l'exploration car le DataFrame n'a pas √©t√© charg√© correctement √† l'√©tape 1.")
```

```python
# √âtape 2.5 : Examiner les valeurs des colonnes qui devraient √™tre num√©riques mais sont "object"

if df_loaded:
    colonnes_a_examiner = [
        'Taux_remplissage (en %)',
        'Temp√©rature (en ¬∞C)',
        'Tension (V)',
        'Intensit√© avant entretien (en A)',
        'Intensit√© apr√®s entretien (en A)',
        'Lin√©aire (1 / 0)'
    ]

    print("\n--- Examen des valeurs dans les colonnes num√©riques potentielles (actuellement object) ---")

    for col in colonnes_a_examiner:
        if col in df.columns:
            print(f"\nPremi√®res valeurs uniques dans '{col}' :")
            # Afficher quelques valeurs uniques pour voir le format
            # Utiliser .astype(str) au cas o√π il y aurait des valeurs non-string (comme NaN)
            unique_values = df[col].astype(str).unique()
            if len(unique_values) > 10: # Si trop de valeurs uniques, n'en afficher que quelques-unes
                print(unique_values[:10])
                print("...")
            else:
                print(unique_values)
        else:
            print(f"La colonne '{col}' n'existe pas dans le DataFrame.")

else:
    print("\nLe DataFrame n'a pas √©t√© charg√©. Impossible d'examiner les colonnes.")
```

```python
# √âtape 2.6 : G√©rer les valeurs manquantes - Supprimer la colonne 'Information_Frigo'

if df_loaded:
    colonne_a_supprimer = 'Information_Frigo'

    if colonne_a_supprimer in df.columns:
        df = df.drop(columns=[colonne_a_supprimer])
        print(f"\n‚úÖ Colonne '{colonne_a_supprimer}' supprim√©e car elle contenait beaucoup de valeurs manquantes.")
        print(f"Nouvelles dimensions du DataFrame : {df.shape[0]} lignes et {df.shape[1]} colonnes.")
    else:
        print(f"\nLa colonne '{colonne_a_supprimer}' n'a pas √©t√© trouv√©e.")

    # On peut aussi rev√©rifier rapidement les manquants pour voir le r√©sultat
    print("\nNombre de valeurs manquantes par colonne apr√®s suppression :")
    print(df.isnull().sum())

else:
    print("\nLe DataFrame n'a pas √©t√© charg√©. Impossible de supprimer la colonne.")
```

```python
# √âtape 2.7 : Nettoyage et conversion des colonnes num√©riques (actuellement object)

if df_loaded:
    print("\n--- D√©but du nettoyage et de la conversion des colonnes num√©riques ---")

    # Copie du DataFrame pour √©viter les avertissements de SettingWithCopyWarning
    df_cleaned = df.copy()

    # Colonnes √† convertir et leur logique de nettoyage
    conversion_rules = {
        'Taux_remplissage (en %)': {
            'clean': lambda x: x, # Pas de nettoyage de caract√®res sp√©cifique, juste conversion
            'dtype': float,
            'errors': 'coerce' # Convertit les erreurs en NaN
        },
        'Temp√©rature (en ¬∞C)': {
            'clean': lambda x: str(x).replace('T=', '').replace('¬∞C', '').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Tension (V)': {
            'clean': lambda x: str(x).replace(' V', '').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Intensit√© avant entretien (en A)': {
            'clean': lambda x: str(x).replace(' A', '').replace(',', '.').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Intensit√© apr√®s entretien (en A)': {
             'clean': lambda x: str(x).replace(' A', '').replace(',', '.').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Lin√©aire (1 / 0)': {
            'clean': lambda x: str(x).strip(), # Nettoyage initial, g√©rer '/' apr√®s
            'dtype': int, # Tenter de convertir en int apr√®s gestion de '/'
            'errors': 'coerce' # Convertit les erreurs (comme '/') en NaN initialement
        }
    }

    for col, rules in conversion_rules.items():
        if col in df_cleaned.columns:
            print(f"Tentative de nettoyage et conversion de '{col}'...")
            try:
                # Appliquer la fonction de nettoyage
                df_cleaned[col] = df_cleaned[col].apply(rules['clean'])

                # Pour 'Lin√©aire (1 / 0)', remplacer '/' ou d'autres valeurs non-0/1 par NaN avant conversion en int
                if col == 'Lin√©aire (1 / 0)':
                     # Remplacer sp√©cifiquement '/' par NaN (car 'coerce' g√®re les autres erreurs de conversion int)
                    df_cleaned[col] = df_cleaned[col].replace('/', pd.NA)
                    # Ensuite, convertir en num√©rique, 'coerce' transformera ce qui n'est pas 0 ou 1 en NaN
                    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
                    # Apr√®s conversion en numeric (float avec NaN), on peut le convertir en Int64 (permet NaN)
                    # Note: 'Int64' avec un I majuscule g√®re les NaN pour les entiers
                    df_cleaned[col] = df_cleaned[col].astype('Int64')

                else:
                    # Convertir en type num√©rique sp√©cifi√©
                    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors=rules['errors'])
                    # Pour les floats, on peut aussi convertir en type qui g√®re les NaN (optionnel, mais propre)
                    if rules['dtype'] == float:
                        df_cleaned[col] = df_cleaned[col].astype('float64') # Utiliser float64

                print(f"‚úÖ Colonne '{col}' nettoy√©e et convertie.")

            except Exception as e:
                print(f"‚ùå Erreur lors du nettoyage/conversion de '{col}': {e}")
        else:
            print(f"La colonne '{col}' n'existe pas dans le DataFrame.")

    # Afficher les types de donn√©es apr√®s conversion
    print("\nTypes de donn√©es apr√®s conversion :")
    print(df_cleaned.dtypes)

    # Afficher le nombre de valeurs manquantes apr√®s conversion (les erreurs deviennent NaN)
    print("\nNombre de valeurs manquantes par colonne apr√®s conversion :")
    print(df_cleaned.isnull().sum())

    # Mettre √† jour le DataFrame principal si le nettoyage a r√©ussi
    df = df_cleaned

else:
    print("\nLe DataFrame n'a pas √©t√© charg√©. Impossible de nettoyer et convertir les colonnes.")
```

```python
# √âtape 2.8 : G√©rer les valeurs manquantes restantes

if df_loaded:
    print("\n--- D√©but de la gestion des valeurs manquantes restantes ---")

    # Imputation pour les colonnes num√©riques avec la m√©diane
    numeric_cols_to_impute = [
        'Taux_remplissage (en %)',
        'Temp√©rature (en ¬∞C)',
        'Lin√©aire (1 / 0)',
        'Tension (V)',
        'Intensit√© avant entretien (en A)',
        'Intensit√© apr√®s entretien (en A)'
    ]

    for col in numeric_cols_to_impute:
        if col in df.columns:
            # Calculer la m√©diane en ignorant les valeurs NaN
            median_value = df[col].median()
            # Remplacer les valeurs NaN par la m√©diane
            df[col].fillna(median_value, inplace=True)
            print(f"‚úÖ Valeurs manquantes de '{col}' imput√©es avec la m√©diane ({median_value:.2f}).")
        else:
             print(f"La colonne num√©rique '{col}' n'existe pas dans le DataFrame.")


    # Imputation pour les colonnes object avec "Inconnu"
    object_cols_to_impute = [
        'Nom_Client',
        'Nom_PDV',
        'Tel_BARMAN',
        'Localisation',
        'AF / NF',
        'Serial_Number',
        'Tag_Number'
    ]

    for col in object_cols_to_impute:
        if col in df.columns:
            df[col].fillna("Inconnu", inplace=True)
            print(f"‚úÖ Valeurs manquantes de '{col}' imput√©es avec 'Inconnu'.")
        else:
             print(f"La colonne object '{col}' n'existe pas dans le DataFrame.")


    print("\nNombre de valeurs manquantes apr√®s imputation :")
    print(df.isnull().sum())

else:
    print("\nLe DataFrame n'a pas √©t√© charg√©. Impossible de g√©rer les valeurs manquantes restantes.")
```

```python
# √âtape 3 : Analyse Descriptive de Base

if df_loaded:
    print("\n--- D√©but de l'analyse descriptive ---")

    # 3.1 Statistiques descriptives pour les colonnes num√©riques
    print("\nStatistiques descriptives des colonnes num√©riques :")
    # Utiliser display() pour une meilleure mise en forme dans Colab
    display(df.describe())

    # 3.2 Fr√©quence des valeurs pour certaines colonnes cat√©gorielles
    # Choisis les colonnes cat√©gorielles qui t'int√©ressent le plus pour cette analyse
    categorical_cols_to_analyze = [
        'Division',
        'Secteur',
        'Ville',
        'Type_Frigo',
        'AF / NF',
        'Branding',
        'Securit√© ( Disjoncteur / R√©gulateur )',
        'Eclairage (O / N )',
        'Purge du circuit d\'√©valuation des eaux',
        'Soufflage des parties actives √† l\'air'
    ]

    print("\nFr√©quence des valeurs pour les colonnes cat√©gorielles s√©lectionn√©es :")

    for col in categorical_cols_to_analyze:
        if col in df.columns:
            print(f"\n--- Fr√©quence pour '{col}' ---")
            # Utiliser value_counts() pour compter les occurrences de chaque valeur unique
            # Normaliser=True pour obtenir les pourcentages
            # dropna=False pour inclure les valeurs manquantes si elles existaient encore (ici elles sont imput√©es)
            print(df[col].value_counts(dropna=False))
        else:
            print(f"La colonne cat√©gorielle '{col}' n'existe pas dans le DataFrame.")

else:
    print("\nLe DataFrame n'a pas √©t√© charg√©. Impossible de proc√©der √† l'analyse descriptive.")
```

```python
# √âtape 2.9 (Extension du nettoyage) : G√©rer les valeurs '/' dans les colonnes cat√©gorielles

if df_loaded:
    print("\n--- D√©but de la gestion des valeurs '/' dans les colonnes cat√©gorielles ---")

    categorical_cols_with_slash = [
        'Type_Frigo',
        'AF / NF',
        'Securit√© ( Disjoncteur / R√©gulateur )',
        'Eclairage (O / N )',
        "Purge du circuit d'√©valuation des eaux",
        "Soufflage des parties actives √† l'air"
    ]

    for col in categorical_cols_with_slash:
        if col in df.columns:
            # Remplacer la valeur '/' par 'Non renseign√©'
            # Correction pour √©viter FutureWarning
            df[col] = df[col].replace('/', 'Non renseign√©')
            print(f"‚úÖ Valeurs '/' dans '{col}' remplac√©es par 'Non renseign√©'.")
        else:
            print(f"La colonne cat√©gorielle '{col}' n'existe pas dans le DataFrame.")

    print("\nV√©rification des fr√©quences apr√®s remplacement des '/' :")
    # Rev√©rifier les fr√©quences pour confirmer le changement
    for col in categorical_cols_with_slash:
         if col in df.columns:
            print(f"\n--- Fr√©quence mise √† jour pour '{col}' ---")
            print(df[col].value_counts(dropna=False))
         else:
            print(f"La colonne cat√©gorielle '{col}' n'existe pas dans le DataFrame.")

else:
    print("\nLe DataFrame n'a pas √©t√© charg√©. Impossible de g√©rer les valeurs '/'.")
```
