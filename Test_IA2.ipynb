```python
# Étape 2.6 : Gérer les valeurs manquantes - Supprimer la colonne 'Information_Frigo'

if df_loaded:
    colonne_a_supprimer = 'Information_Frigo'

    if colonne_a_supprimer in df.columns:
        df = df.drop(columns=[colonne_a_supprimer])
        print(f"\n✅ Colonne '{colonne_a_supprimer}' supprimée car elle contenait beaucoup de valeurs manquantes.")
        print(f"Nouvelles dimensions du DataFrame : {df.shape[0]} lignes et {df.shape[1]} colonnes.")
    else:
        print(f"\nLa colonne '{colonne_a_supprimer}' n'a pas été trouvée.")

    # On peut aussi revérifier rapidement les manquants pour voir le résultat
    print("\nNombre de valeurs manquantes par colonne après suppression :")
    print(df.isnull().sum())

else:
    print("\nLe DataFrame n'a pas été chargé. Impossible de supprimer la colonne.")
```

```python
# Étape 2.7 : Nettoyage et conversion des colonnes numériques (actuellement object)

if df_loaded:
    print("\n--- Début du nettoyage et de la conversion des colonnes numériques ---")

    # Copie du DataFrame pour éviter les avertissements de SettingWithCopyWarning
    df_cleaned = df.copy()

    # Colonnes à convertir et leur logique de nettoyage
    conversion_rules = {
        'Taux_remplissage (en %)': {
            'clean': lambda x: x, # Pas de nettoyage de caractères spécifique, juste conversion
            'dtype': float,
            'errors': 'coerce' # Convertit les erreurs en NaN
        },
        'Température (en °C)': {
            'clean': lambda x: str(x).replace('T=', '').replace('°C', '').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Tension (V)': {
            'clean': lambda x: str(x).replace(' V', '').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Intensité avant entretien (en A)': {
            'clean': lambda x: str(x).replace(' A', '').replace(',', '.').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Intensité après entretien (en A)': {
             'clean': lambda x: str(x).replace(' A', '').replace(',', '.').strip(),
            'dtype': float,
            'errors': 'coerce'
        },
        'Linéaire (1 / 0)': {
            'clean': lambda x: str(x).strip(), # Nettoyage initial, gérer '/' après
            'dtype': int, # Tenter de convertir en int après gestion de '/'
            'errors': 'coerce' # Convertit les erreurs (comme '/') en NaN initialement
        }
    }

    for col, rules in conversion_rules.items():
        if col in df_cleaned.columns:
            print(f"Tentative de nettoyage et conversion de '{col}'...")
            try:
                # Appliquer la fonction de nettoyage
                df_cleaned[col] = df_cleaned[col].apply(rules['clean'])

                # Pour 'Linéaire (1 / 0)', remplacer '/' ou d'autres valeurs non-0/1 par NaN avant conversion en int
                if col == 'Linéaire (1 / 0)':
                     # Remplacer spécifiquement '/' par NaN (car 'coerce' gère les autres erreurs de conversion int)
                    df_cleaned[col] = df_cleaned[col].replace('/', pd.NA)
                    # Ensuite, convertir en numérique, 'coerce' transformera ce qui n'est pas 0 ou 1 en NaN
                    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
                    # Après conversion en numeric (float avec NaN), on peut le convertir en Int64 (permet NaN)
                    # Note: 'Int64' avec un I majuscule gère les NaN pour les entiers
                    df_cleaned[col] = df_cleaned[col].astype('Int64')

                else:
                    # Convertir en type numérique spécifié
                    df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors=rules['errors'])
                    # Pour les floats, on peut aussi convertir en type qui gère les NaN (optionnel, mais propre)
                    if rules['dtype'] == float:
                        df_cleaned[col] = df_cleaned[col].astype('float64') # Utiliser float64

                print(f"✅ Colonne '{col}' nettoyée et convertie.")

            except Exception as e:
                print(f"❌ Erreur lors du nettoyage/conversion de '{col}': {e}")
        else:
            print(f"La colonne '{col}' n'existe pas dans le DataFrame.")

    # Afficher les types de données après conversion
    print("\nTypes de données après conversion :")
    print(df_cleaned.dtypes)

    # Afficher le nombre de valeurs manquantes après conversion (les erreurs deviennent NaN)
    print("\nNombre de valeurs manquantes par colonne après conversion :")
    print(df_cleaned.isnull().sum())

    # Mettre à jour le DataFrame principal si le nettoyage a réussi
    df = df_cleaned

else:
    print("\nLe DataFrame n'a pas été chargé. Impossible de nettoyer et convertir les colonnes.")
```

```python
# Étape 2.8 : Gérer les valeurs manquantes restantes

if df_loaded:
    print("\n--- Début de la gestion des valeurs manquantes restantes ---")

    # Imputation pour les colonnes numériques avec la médiane
    numeric_cols_to_impute = [
        'Taux_remplissage (en %)',
        'Température (en °C)',
        'Linéaire (1 / 0)',
        'Tension (V)',
        'Intensité avant entretien (en A)',
        'Intensité après entretien (en A)'
    ]

    for col in numeric_cols_to_impute:
        if col in df.columns:
            # Calculer la médiane en ignorant les valeurs NaN
            median_value = df[col].median()
            # Remplacer les valeurs NaN par la médiane
            df[col].fillna(median_value, inplace=True)
            print(f"✅ Valeurs manquantes de '{col}' imputées avec la médiane ({median_value:.2f}).")
        else:
             print(f"La colonne numérique '{col}' n'existe pas dans le DataFrame.")


    # Imputation pour les colonnes object avec "Inconnu"
    object_cols_to_impute = [
        'Nom_Client',
        'Nom_PDV',
        'Tel_BARMAN',
        'Localisation',
        'AF / NF',
        'Serial_Number',
        'Tag_Number'
    ]

    for col in object_cols_to_impute:
        if col in df.columns:
            df[col].fillna("Inconnu", inplace=True)
            print(f"✅ Valeurs manquantes de '{col}' imputées avec 'Inconnu'.")
        else:
             print(f"La colonne object '{col}' n'existe pas dans le DataFrame.")


    print("\nNombre de valeurs manquantes après imputation :")
    print(df.isnull().sum())

else:
    print("\nLe DataFrame n'a pas été chargé. Impossible de gérer les valeurs manquantes restantes.")
```

```python
# Étape 2.9 (Extension du nettoyage) : Gérer les valeurs '/' dans les colonnes catégorielles

if df_loaded:
    print("\n--- Début de la gestion des valeurs '/' dans les colonnes catégorielles ---")

    categorical_cols_with_slash = [
        'Type_Frigo',
        'AF / NF',
        'Securité ( Disjoncteur / Régulateur )',
        'Eclairage (O / N )',
        "Purge du circuit d'évaluation des eaux",
        "Soufflage des parties actives à l'air"
    ]

    for col in categorical_cols_with_slash:
        if col in df.columns:
            # Remplacer la valeur '/' par 'Non renseigné'
            # Correction pour éviter FutureWarning
            df[col] = df[col].replace('/', 'Non renseigné')
            print(f"✅ Valeurs '/' dans '{col}' remplacées par 'Non renseigné'.")
        else:
            print(f"La colonne catégorielle '{col}' n'existe pas dans le DataFrame.")

    print("\nVérification des fréquences après remplacement des '/' :")
    # Revérifier les fréquences pour confirmer le changement
    for col in categorical_cols_with_slash:
         if col in df.columns:
            print(f"\n--- Fréquence mise à jour pour '{col}' ---")
            print(df[col].value_counts(dropna=False))
         else:
            print(f"La colonne catégorielle '{col}' n'existe pas dans le DataFrame.")

else:
    print("\nLe DataFrame n'a pas été chargé. Impossible de gérer les valeurs '/'.")
```

```python
# Étape 6 : Création de la colonne Statut Post-Entretien

if df_loaded:
    print("\n--- Création de la colonne Statut Post-Entretien ---")

    # Définir les seuils basés sur ton analyse
    seuil_investigation = 2.0
    seuil_echec_resistance = 3.1
    seuil_maintenance_preventive_bas = 2.4
    seuil_maintenance_preventive_haut = 2.6

    # Créer une liste de conditions et de valeurs correspondantes
    conditions = [
        (df['Intensité après entretien (en A)'] < seuil_investigation),
        (df['Intensité après entretien (en A)'] > seuil_echec_resistance),
        (df['Intensité après entretien (en A)'] >= seuil_maintenance_preventive_bas) & (df['Intensité après entretien (en A)'] <= seuil_maintenance_preventive_haut)
    ]

    choices = [
        'Investigation_defaillance',
        'Entretien_renforce',
        'Maintenance_preventive'
    ]

    # Utiliser np.select pour appliquer les conditions
    # np.select permet de définir des cas multiples, le dernier argument est la valeur par défaut
    import numpy as np
    df['Statut_Post_Entretien'] = np.select(conditions, choices, default='Surveillance_renforcee')

    print("✅ Colonne 'Statut_Post_Entretien' créée.")

    # Afficher les fréquences pour vérifier la répartition
    print("\nRépartition des Statuts Post-Entretien :")
    print(df['Statut_Post_Entretien'].value_counts())

else:
    print("\nLe DataFrame n'a pas été chargé. Impossible de créer la colonne Statut Post-Entretien.")
```

```python
# Étape 7 : Analyse de la Répartition du Statut Post-Entretien par Type de Frigo

if df_loaded:
    print("\n--- Analyse de la Répartition du Statut Post-Entretien par Type de Frigo ---")

    # Réutiliser la liste des 5 types de frigo les plus fréquents de l'étape 5
    top_frigo_types = df['Type_Frigo'].value_counts().nlargest(5).index.tolist()
    print(f"\nAnalyse de la répartition des statuts pour les 5 types de frigo les plus fréquents : {top_frigo_types}")

    # Filtrer le DataFrame pour inclure seulement ces types
    df_top_types = df[df['Type_Frigo'].isin(top_frigo_types)].copy()

    # Analyser la répartition du statut par Type de Frigo
    print("\nRépartition du Statut Post-Entretien par Type de Frigo :")
    # Utiliser crosstab pour créer un tableau de contingence
    statut_par_type = pd.crosstab(df_top_types['Type_Frigo'], df_top_types['Statut_Post_Entretien'])

    # Ajouter les totaux par ligne et par colonne
    statut_par_type['Total'] = statut_par_type.sum(axis=1)
    statut_par_type.loc['Total Général'] = statut_par_type.sum()

    display(statut_par_type)

    # Optionnel : Afficher les pourcentages par Type de Frigo
    print("\nRépartition en Pourcentage du Statut Post-Entretien par Type de Frigo (par ligne) :")
    statut_par_type_pct = pd.crosstab(df_top_types['Type_Frigo'], df_top_types['Statut_Post_Entretien'], normalize='index') * 100
    display(statut_par_type_pct)

else:
    print("\nLe DataFrame n'a pas été chargé. Impossible d'analyser la répartition par Type de Frigo.")
```

```python
# Étape 10 : Sélection des Features et de la Cible

if df_loaded:
    print("\n--- Sélection des Features et de la Cible ---")

    # Variable Cible (y)
    target_column = 'Statut_Post_Entretien'
    if target_column in df.columns:
        y = df[target_column]
        print(f"✅ Variable cible '{target_column}' sélectionnée.")
    else:
        print(f"❌ Erreur : La colonne cible '{target_column}' n'existe pas dans le DataFrame.")
        y = None # Assigner None si la cible n'est pas trouvée

    # Variables Prédictives (X)
    # Exclure la cible et les colonnes non pertinentes/calculées après l'entretien
    columns_to_exclude = [
        target_column,
        'Nom_Client',
        'Nom_PDV',
        'Tel_BARMAN',
        'Localisation',
        'Serial_Number',
        'Tag_Number',
        'Intensité après entretien (en A)', # Exclure car calculée après l'entretien
        'Delta_Intensite (A)',              # Exclure car calculée après l'entretien
        'Efficacite_Entretien (%)'          # Exclure car calculée après l'entretien
    ]

    # Sélectionner toutes les autres colonnes comme features potentielles
    X = df.drop(columns=columns_to_exclude, errors='ignore') # errors='ignore' pour ne pas planter si une colonne n'existe pas

    print(f"✅ Variables prédictives sélectionnées (X) avec {X.shape[1]} colonnes.")
    print("Les premières lignes de X :")
    display(X.head())
    print("\nLes premières valeurs de y :")
    print(y.head())


else:
    print("\nLe DataFrame n'a pas été chargé. Impossible de sélectionner les features et la cible.")
```

```python
# Étape 11 : Encodage des Variables Catégorielles (Correction finale)

if df_loaded and X is not None and y is not None:
    print("\n--- Début de l'Encodage des Variables Catégorielles ---")

    from sklearn.preprocessing import OneHotEncoder
    from sklearn.compose import ColumnTransformer
    # from sklearn.pipeline import Pipeline # Plus nécessaire directement ici
    import pandas as pd
    import numpy as np # Assurons-nous que numpy est importé

    # Identifier les colonnes numériques et catégorielles dans X
    numeric_features = X.select_dtypes(include=np.number).columns.tolist()
    categorical_features = X.select_dtypes(include='object').columns.tolist()

    # Liste des colonnes à IGNORER complètement par le préprocesseur
    columns_to_ignore = ['Date']
    # Assurons-nous que 'Date' est bien retirée des autres listes si elle y était par erreur
    numeric_features = [col for col in numeric_features if col not in columns_to_ignore]
    categorical_features = [col for col in categorical_features if col not in columns_to_ignore]


    print(f"Colonnes numériques à traiter (incluses) : {numeric_features}")
    print(f"Colonnes catégorielles à encoder (incluses) : {categorical_features}")
    print(f"Colonnes à ignorer : {columns_to_ignore}")


    # Créer le préprocesseur (ColumnTransformer)
    # Il appliquera différentes transformations aux colonnes numériques et catégorielles
    # Correction : un seul argument 'transformers'
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
            ('num', 'passthrough', numeric_features), # Laisse passer les numériques
            ('ignore', 'drop', columns_to_ignore) # Ignore explicitement la Date
        ]
    )


    # Appliquer le préprocesseur à X
    # X_processed sera un tableau NumPy
    X_processed = preprocessor.fit_transform(X)

    print("\n✅ Encodage et prétraitement appliqués.")
    print(f"Nouvelles dimensions de X après encodage : {X_processed.shape}")
    # Les dimensions devraient augmenter en raison des colonnes créées par le one-hot encoding

    # Optionnel : Si tu veux récupérer les noms des colonnes encodées
    # C'est un peu complexe avec ColumnTransformer et OneHotEncoder
    # Les noms des features après OneHotEncoder sont accessibles via get_feature_names_out()
    # Mais il faut gérer l'ordre des colonnes après le transformer.
    # Pour l'instant, contentons-nous de la forme NumPy.

else:
    print("\nLe DataFrame X ou y n'est pas correctement chargé. Impossible de procéder à l'encodage.")
```

```python
# Étape 12 : Division des Données (Train/Test Split)

if df_loaded and X_processed is not None and y is not None:
    print("\n--- Début de la Division des Données ---")

    from sklearn.model_selection import train_test_split

    # Diviser les données en ensembles d'entraînement et de test
    # test_size: la proportion de données à inclure dans l'ensemble de test (par exemple, 20%)
    # random_state: permet de reproduire la même division à chaque exécution
    # stratify: assure que la proportion de chaque classe de la variable cible (y) est la même dans les ensembles d'entraînement et de test.
    # C'est important pour les problèmes de classification, surtout si certaines classes sont rares.
    X_train, X_test, y_train, y_test = train_test_split(
        X_processed, # Tes features encodées
        y,             # Ta variable cible (Statut_Post_Entretien)
        test_size=0.2, # Utilise 20% des données pour le test
        random_state=42, # Un nombre arbitraire pour la reproductibilité
        stratify=y     # IMPORTANT pour la classification multi-classes
    )

    print(f"✅ Données divisées en ensembles d'entraînement et de test.")
    print(f"Dimensions de l'ensemble d'entraînement (X_train) : {X_train.shape}")
    print(f"Dimensions de l'ensemble de test (X_test) : {X_test.shape}")
    print(f"Dimensions de la cible d'entraînement (y_train) : {y_train.shape}")
    print(f"Dimensions de la cible de test (y_test) : {y_test.shape}")

    # Optionnel : vérifier la répartition des classes dans y_train et y_test (devrait être similaire)
    print("\nRépartition des classes dans y_train :")
    print(y_train.value_counts(normalize=True))
    print("\nRépartition des classes dans y_test :")
    print(y_test.value_counts(normalize=True))


else:
    print("\nLes données (X_processed ou y) ne sont pas correctement chargées. Impossible de diviser les données.")
```

```python
# Étape 13 : Entraînement et Évaluation d'un Modèle de Classification (Random Forest)

if df_loaded and X_train is not None and X_test is not None and y_train is not None and y_test is not None:
    print("\n--- Début de l'Entraînement et de l'Évaluation du Modèle ---")

    from sklearn.ensemble import RandomForestClassifier
    from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

    # Créer une instance du modèle Random Forest
    # n_estimators: le nombre d'arbres dans la forêt (100 est une bonne valeur de départ)
    # random_state: pour la reproductibilité
    model = RandomForestClassifier(n_estimators=100, random_state=42)

    # Entraîner le modèle sur l'ensemble d'entraînement
    print("Entraînement du modèle...")
    model.fit(X_train, y_train)
    print("✅ Modèle entraîné.")

    # Faire des prédictions sur l'ensemble de test
    print("\nRéalisation des prédictions sur l l'ensemble de test...")
    y_pred = model.predict(X_test)
    print("✅ Prédictions réalisées.")

    # Évaluer la performance du modèle
    print("\n--- Évaluation du Modèle ---")

    # Précision globale (Accuracy)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Précision (Accuracy) sur l'ensemble de test : {accuracy:.4f}")

    # Matrice de confusion
    print("\nMatrice de Confusion :")
    print(confusion_matrix(y_test, y_pred))
    # Note: La matrice de confusion est un tableau où chaque ligne représente les vraies classes et chaque colonne les classes prédites.
    # L'élément (i, j) est le nombre d'échantillons de la classe i qui ont été prédits comme appartenant à la classe j.
    # Les éléments sur la diagonale (haut gauche à bas droite) sont les prédictions correctes.

    # Rapport de classification (Precision, Recall, F1-score par classe)
    print("\nRapport de Classification :")
    print(classification_report(y_test, y_pred))
    # Note:
    # Precision: Pour une classe donnée, c'est la proportion de prédictions positives qui étaient réellement positives. (True Positives / (True Positives + False Positives))
    # Recall (Sensitivity): Pour une classe donnée, c'est la proportion d'exemples positifs réels qui ont été correctement identifiés. (True Positives / (True Positives + False Negatives))
    # F1-score: La moyenne harmonique de la précision et du rappel. C'est une bonne métrique globale, surtout avec des classes déséquilibrées.
    # Support: Le nombre réel d'occurrences de la classe dans l'ensemble de test.

else:
    print("\nLes ensembles d'entraînement/test ne sont pas correctement préparés. Impossible d'entraîner le modèle.")
```

```python
# Étape 14 : Analyse de l'Importance des Features

if df_loaded and model is not None and X is not None and preprocessor is not None:
    print("\n--- Analyse de l'Importance des Features ---")

    # Récupérer les noms des features après le préprocesseur
    # C'est un peu complexe avec ColumnTransformer, mais nécessaire pour interpréter l'importance
    # On va essayer d'utiliser get_feature_names_out() qui est la méthode recommandée

    try:
        # get_feature_names_out() est disponible sur les versions récentes de sklearn
        feature_names = preprocessor.get_feature_names_out()
        print("✅ Noms des features prétraitées récupérés.")
    except Exception as e:
        print(f"❌ Erreur lors de la récupération des noms des features : {e}")
        print("Utilisation des noms de colonnes d'origine comme fallback (peut être moins précis pour les encodées).")
        # Fallback si get_feature_names_out n'est pas disponible ou échoue
        # Cela mélangera les noms originaux avec les noms des colonnes one-hot encodées (si on arrive à les récupérer)
        # C'est une approximation, mais mieux que rien.
        # Une meilleure approche serait de reconstruire les noms manuellement, mais c'est plus complexe.
        # Pour l'instant, on va essayer d'utiliser une méthode qui marche avec OneHotEncoder
        ohe_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
        passthrough_features = X[numeric_features].columns.tolist()
        feature_names = np.concatenate([ohe_feature_names, passthrough_features])


    # Récupérer l'importance des features depuis le modèle entraîné
    importance = model.feature_importances_

    # Créer un DataFrame pour afficher l'importance des features
    feature_importance_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': importance
    })

    # Trier les features par importance (décroissante)
    feature_importance_df = feature_importance_df.sort_values('Importance', ascending=False)

    # Afficher les features les plus importantes (par exemple, les 20 premières)
    print("\nFeatures les plus importantes pour la prédiction du Statut Post-Entretien :")
    display(feature_importance_df.head(20))

    # Optionnel : Visualiser l'importance des features (par exemple, les 10 premières)
    print("\n--- Visualisation des 10 Features les plus importantes ---")
    plt.figure(figsize=(10, 7))
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(10))
    plt.title('Importance des Features pour la Prédiction du Statut Post-Entretien')
    plt.xlabel('Importance (basée sur Mean Decrease in Impurity)')
    plt.ylabel('Feature')
    plt.tight_layout()
    plt.show()

else:
    print("\nLe DataFrame, le modèle ou le préprocesseur ne sont pas correctement chargés. Impossible d'analyser l'importance des features.")
```

```python
# Étape 15 : Optimisation des Hyperparamètres du Modèle

if df_loaded and X_train is not None and y_train is not None:
    print("\n--- Début de l'Optimisation des Hyperparamètres (GridSearchCV) ---")

    from sklearn.model_selection import GridSearchCV
    from sklearn.ensemble import RandomForestClassifier

    # Définir la grille de paramètres à tester
    # Ces valeurs sont des exemples, ajuste-les en fonction de ton temps de calcul et de tes besoins
    param_grid = {
        'n_estimators': [50, 100, 200], # Nombre d'arbres
        'max_depth': [None, 10, 20, 30], # Profondeur maximale de l'arbre
        'min_samples_split': [2, 5, 10], # Nombre minimum d'échantillons requis pour fendre un nœud interne
        'min_samples_leaf': [1, 2, 4],   # Nombre minimum d'échantillons requis pour être à un nœud feuille
        'bootstrap': [True, False]      # Méthode d'échantillonnage des données (avec ou sans remplacement)
    }

    # Créer une instance du modèle Random Forest de base
    rf = RandomForestClassifier(random_state=42)

    # Créer l'objet GridSearchCV
    # estimator: le modèle à optimiser
    # param_grid: la grille de paramètres à explorer
    # cv: le nombre de plis pour la validation croisée (par exemple, 5)
    # scoring: la métrique d'évaluation (par exemple, 'accuracy' pour la précision)
    # n_jobs: nombre de cœurs de processeur à utiliser (-1 utilise tous les cœurs disponibles)
    grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                               cv=5, scoring='accuracy', n_jobs=-1)

    # Lancer la recherche sur grille sur les données d'entraînement
    print("Lancement de la recherche sur grille (cela peut prendre du temps)...")
    grid_search.fit(X_train, y_train)
    print("✅ Recherche sur grille terminée.")

    # Afficher les meilleurs paramètres trouvés
    print("\nMeilleurs paramètres trouvés :")
    print(grid_search.best_params_)

    # Afficher le meilleur score obtenu avec ces paramètres
    print("\nMeilleur score (Accuracy) de validation croisée :")
    print(grid_search.best_score_)

    # Le meilleur modèle est maintenant accessible via grid_search.best_estimator_
    best_model = grid_search.best_estimator_

    # Évaluer le meilleur modèle sur l'ensemble de test (pour une évaluation finale non biaisée)
    print("\n--- Évaluation du meilleur modèle sur l'ensemble de test ---")
    y_pred_best = best_model.predict(X_test)

    accuracy_best = accuracy_score(y_test, y_pred_best)
    print(f"Précision (Accuracy) du meilleur modèle sur l'ensemble de test : {accuracy_best:.4f}")

    print("\nRapport de Classification du meilleur modèle :")
    print(classification_report(y_test, y_pred_best))

else:
    print("\nLes ensembles d'entraînement/test ne sont pas correctement préparés pour l'optimisation.")
```

```python
# Étape 16 : Analyse des Erreurs de Prédiction

if df_loaded and best_model is not None and X_test is not None and y_test is not None:
    print("\n--- Début de l'Analyse des Erreurs de Prédiction ---")

    # Créer un DataFrame pour comparer les vraies valeurs et les prédictions
    # C'est plus facile si y_test est sous forme de Series Pandas
    if isinstance(y_test, np.ndarray):
        y_test_series = pd.Series(y_test, name='True_Status')
    else:
         y_test_series = y_test.reset_index(drop=True) # Assurez-vous d'avoir un index propre


    predictions_df = pd.DataFrame({
        'True_Status': y_test_series,
        'Predicted_Status': y_pred_best
    })

    # Identifier les lignes où la prédiction est incorrecte
    errors_df = predictions_df[predictions_df['True_Status'] != predictions_df['Predicted_Status']]

    print(f"\nNombre total d'erreurs de prédiction sur l'ensemble de test : {len(errors_df)}")

    if not errors_df.empty:
        print("\nExemples d'erreurs de prédiction :")
        display(errors_df.head())

        # Analyser la matrice de confusion plus en détail
        # Afficher la matrice de confusion avec les noms des classes
        from sklearn.metrics import confusion_matrix
        cm = confusion_matrix(y_test, y_pred_best, labels=best_model.classes_)
        cm_df = pd.DataFrame(cm, index=best_model.classes_, columns=best_model.classes_)

        print("\nMatrice de Confusion avec labels :")
        display(cm_df)

        # Analyser les types d'erreurs par classe
        print("\nAnalyse des types d'erreurs (Faux Positifs et Faux Négatifs par classe) :")
        # Un peu plus complexe à obtenir directement, mais on peut déduire de la matrice de confusion
        # Pour chaque classe (i), les faux négatifs sont la somme de la ligne i (sauf la diagonale)
        # Pour chaque classe (j), les faux positifs sont la somme de la colonne j (sauf la diagonale)

        for i, true_class in enumerate(best_model.classes_):
            fn = cm[i, :].sum() - cm[i, i] # Somme de la ligne - élément diagonal
            fp = cm[:, i].sum() - cm[i, i] # Somme de la colonne - élément diagonal
            print(f"Classe '{true_class}': Faux Négatifs = {fn}, Faux Positifs = {fp}")

    else:
        print("Aucune erreur de prédiction trouvée sur l'ensemble de test (ce qui est rare !).")


else:
    print("\nLe modèle ou les données de test ne sont pas prêts pour l'analyse des erreurs.")
```
